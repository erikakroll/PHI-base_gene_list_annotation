```python
import re

def parse_input(file_path):
    with open(file_path, 'r') as file:
        input_string = file.read()
        
    entries = input_string.strip().split('\n')
    parsed_entries = []
    for entry in entries:
        match = re.match(r'^>(\S+?)#(\S+)#(\S+)#(\S+)#(\S+)#(\S+)$', entry)
        if match:
            entry_data = {
                "ID": match.group(1),
                "PHIBASE": match.group(2),
                "Gene_Name": match.group(3),
                "Phenotype": match.group(6)
            }
            parsed_entries.append(entry_data)
    return parsed_entries

def search_for_ids(entries, search_ids):
    search_results = []
    for search_id in search_ids:
        results = [entry for entry in entries if entry['ID'] == search_id]
        search_results.extend(results)
    return search_results

def print_table(data):
    if data:
        print("{:<20} {:<15} {:<15} {:<15} {:<15}".format('Fgram_ID', 'ID', 'PHIBASE', 'Gene_Name', 'Phenotype'))
        for entry in data:
            print("{:<20} {:<15} {:<15} {:<15} {:<15}".format(entry['Fgram_ID'], entry['ID'], entry['PHIBASE'], entry['Gene_Name'], entry['Phenotype']))
    else:
        print("No matching entries found.")

# Parse the input data
file_path_input = "PHI_base_v16.txt"
parsed_entries = parse_input(file_path_input)

# Read search IDs from a text file
file_path_search_ids = "F12_Uniprot_Mapped.txt"
search_ids_map = {}
with open(file_path_search_ids, 'r') as file:
    for line in file:
        columns = line.strip().split()
        if len(columns) == 2:
            search_ids_map[columns[0]] = columns[1]

# Search for IDs and print results in a table
search_results = []
for fgram_id, search_id in search_ids_map.items():
    results = search_for_ids(parsed_entries, [search_id])
    for result in results:
        result['Fgram_ID'] = fgram_id
    search_results.extend(results)

print_table(search_results)
```
