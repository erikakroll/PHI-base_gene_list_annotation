```python
import re

def process_blast_output(file_path):
    with open(file_path, 'r') as file:
        blast_output = file.read()
    # This line splits the blast_output string into individual queries.
    queries = re.split(r'Query=', blast_output)[1:]
    
    # This initializes the table string with column headers for "Query", "PHI", and "Phenotype", separated by tabs.
    table = "Query\tPHI\tGene_Name\tOrganism\tPhenotype\n"

    # This loop iterates over each query block obtained from splitting the BLAST output.
    for query_block in queries:
        # This line extracts the query from the current query block.
        query = re.search(r'(.+)', query_block).group(1).strip()

        # This line finds the first hit within the current query block.
        hit = re.search(r'>(.+?)#(.+?)#(.+?)#(.+?)#(.+?)#(.+?)\n', query_block)
        
        if hit:
            hit_info = hit.groups()
            # Remove whitespace from the phenotype
            phenotype = ' '.join([line.strip() for line in hit_info[-1].split('\n')]).strip()
            # Remove whitespace from the output line
            output_line = hit_info[0].replace(" ", "")
            table += f"{query}\t{hit_info[1]}\t{hit_info[2]}\t{hit_info[4]}\t{phenotype}\n"
    
    return table



# Example usage:
file_path = "F01_NCBI_BLAST_cleaned.txt"




print(process_blast_output(file_path))
```
